```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import zipfile
import os
```

```python
zip_path= "dataset.zip"
extract_to="data/"
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_to)
```

```python
def check_mismatches(image_dir, label_dir, set_name):
    image_files = sorted([f for f in os.listdir(image_dir) if f.endswith(('.jpg', '.png', '.jpeg'))])
    label_files = sorted([f for f in os.listdir(label_dir) if f.endswith('.txt')])

    image_basenames = [os.path.splitext(f)[0] for f in image_files]
    label_basenames = [os.path.splitext(f)[0] for f in label_files]

    images_without_labels = sorted(set(image_basenames) - set(label_basenames))
    labels_without_images = sorted(set(label_basenames) - set(image_basenames))

    print(f"\n[{set_name.upper()} SET]")
    print("Images without labels:", images_without_labels)
    print("Labels without images:", labels_without_images)


# Train set
check_mismatches(
    image_dir='data/Dataset/train/images',
    label_dir='data/Dataset/train/labels',
    set_name='train'
)

# Validation set
check_mismatches(
    image_dir='data/Dataset/valid/images',
    label_dir='data/Dataset/valid/labels',
    set_name='valid'
)
```

```python
from ultralytics import YOLO
model= YOLO("yolov8n.pt")
```

```python
model.train(
    data="data.yaml",
    epochs=10,
    imgsz=640,
    batch=8,
    project="lunar_project",
    name="yolov8_crater_detection",
    verbose=True
)
```

```python
# Load trained model
model = YOLO("lunar_project/yolov8_crater_detection2/weights/best.pt")
metrics = model.val(data="data.yaml")

results = model.predict(source="data/Dataset/valid/images", save=True, conf=0.25)
```

```python
results = model.val(data="data.yaml")

# Get the metrics dictionary
metrics_dict = results.results_dict

# Safely extract values
map_50 = metrics_dict.get('metrics/mAP50(B)', 0.0)
map_5095 = metrics_dict.get('metrics/mAP50-95(B)', 0.0)
precision = metrics_dict.get('metrics/precision(B)', 0.0)
recall = metrics_dict.get('metrics/recall(B)', 0.0)

# Print results
print(f"mAP@0.5: {map_50:.4f}")
print(f"mAP@0.5:0.95: {map_5095:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
```
