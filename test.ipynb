```python
from ultralytics import YOLO
```

```python
model_path=YOLO('lunar_project/yolov8_crater_detection2/weights/best.pt')
model_path.eval()
```

```python
test_dir='test/images'
outputs_dir='test_labels'
output_vis='test_visuals'
os.makedirs(outputs_dir,exist_ok=True)
os.makedirs(output_vis,exist_ok=True)
model= YOLO(model_path)
```

```python
for filename in os.listdir(test_dir):
    if filename.lower().endswith('.jpg'):
        image_path= os.path.join(test_dir,filename)
        results= model(image_path)
        names=results[0].names
        class_counts={}
        for box in results[0].boxes:
            cls_id= int(box.cls[0])
            label= names[cls_id]
            class_counts[label] = class_counts.get(label, 0) +1
        if class_counts:
            predicted_label= max(class_counts, key=class_counts.get)
        else:
            predicted_label='crater'
        base_name= os.path.splitext(filename)[0]
        output_file=os.path.join(outputs_dir,f"{base_name}_label.txt")
        with open(output_file,'w') as f:
            f.write(predicted_label)
        annotated_img=results[0].plot()
        vis_path= os.path.join(output_vis, filename)
        cv2.imwrite(vis_path, annotated_img)
```
